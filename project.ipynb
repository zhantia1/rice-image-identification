In this project I am trying to build a CNN model to recognize and classify images of rice into the different
species of rice that exists in this dataset. This problem is generally useful for any parties interested in
building image recognition software to classify food items. Some examples could be grocery store chains or
rice wholesalers. 

There are 5 different types of rice in the dataset which can be
found here: https://www.kaggle.com/datasets/muratkokludataset/rice-image-dataset

The code can also be found on github here: https://github.com/zhantia1/rice-image-identification

```python
import os
import pandas as pd

path = 'Rice_Image_Dataset'

images = []
labels = []

# loop over the folder and place image path and labels into arrays
for subfolder in os.listdir(path):

    subfolder_path = os.path.join(path, subfolder)
    if not os.path.isdir(subfolder_path):
        continue

    for image_filename in os.listdir(subfolder_path):
        image_path = os.path.join(subfolder_path, image_filename)
        images.append(image_path)
        labels.append(subfolder)

df = pd.DataFrame({'image': images, 'label': labels})
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.countplot(x=df.label)
ax.set_xlabel("Type of Rice")
ax.set_ylabel("Number of samples for each type")

plt.show()
```

```python
# View sample images
def display_sample_images(data, num_samples=5):
    classes = data['label'].unique()
    fig, axs = plt.subplots(len(classes), num_samples, figsize=(15, 10))
    for i, cls in enumerate(classes):
        samples = data[data['label'] == cls].sample(num_samples)['image'].values
        for j, sample in enumerate(samples):
            axs[i, j].imshow(plt.imread(sample))
            axs[i, j].axis('off')
            axs[i, j].set_title(cls)
    plt.tight_layout()
    plt.show()

# Check image sizes
image_sizes = df['image'].apply(lambda x: plt.imread(x).shape[:2])
print("Image Sizes:", image_sizes.unique())

display_sample_images(df)
```

After exploring the data a bit, we will start building out our model. We first split the data
into training and test datasets using a 0.3 split. We create new dataframes for the train
and test splits to use in our data generators.

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.metrics import confusion_matrix

X = df['image']
y = df['label']

# Split the data into train, and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a dataframe for the training data
df_train = pd.DataFrame({'image': X_train, 'label': y_train})

# Create a dataframe for the test data
df_test = pd.DataFrame({'image': X_test, 'label': y_test})
```

Here we instantiate our image data generator as well as define our CNN model. The image data generators
are a data augmentation technicque used to increase the size of the training dataset by transforming the image.
We train the model through 5 epochs.

```python
image_size = (50, 50)
batch_size = 32

datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=45,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_generator = datagen.flow_from_dataframe(
    df_train,
    x_col='image',
    y_col='label',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

test_generator = datagen.flow_from_dataframe(
    df_test,
    x_col='image',
    y_col='label',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)


# Define the CNN architecture using TensorFlow/Keras
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(50, 50, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(5, activation='softmax')  # 5 types of rice
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(train_generator,epochs=5,validation_data=test_generator)
```

After building the model we will want to evaluate the model using some graphs. We first plot
the accuracy comparison through the different epochs.

```python
plt.figure(figsize=(10,6))
plt.plot(history.history['accuracy'], color="#E74C3C", marker='o')
plt.plot(history.history['val_accuracy'], color='#641E16', marker='h')
plt.title('Accuracy comparison between Train and Test Data set', fontsize=15)
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

Next we will plot a loss comparison between Validation and Train Data set.

```python

plt.figure(figsize=(10,6))
plt.plot(history.history['loss'],color="#E74C3C",marker='o')
plt.plot(history.history['val_loss'],color='#641E16',marker='h')
plt.title('Loss comparison between Test and Train Data set',fontsize=15)
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

Another thing we are interested in looking at is the confusion matrix. This shows the
true positives and true negatives of our model.

```python
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = test_generator.classes

conf_matrix = confusion_matrix(y_true_classes, y_pred_classes)

print(conf_matrix)
```

Finally we will evaluate the model using the test data and get the accuracy.

```python
# Evaluate the model on the test data
metrics = model.evaluate(test_generator)
# Print the accuracy of the model
print('Accuracy:', metrics[1])
```
